```python
def trapezio(f,a,b,N):
    
    x = np.linspace(a,b,N+1) # N+1 pontos
    y = f(x)
    y_inter = y[1:N] # todos valores menos o primeiro e o ultimo
    dx = (b - a)/N
    T = (dx)*np.sum(y_inter)+(dx/2)*(y[0]+y[N])
    return T
```

```python
# A
x = np.linspace(1.0,1.6,4)
y = np.log(x)
I = np.trapz(y,x)
print(I)

defTraz = trapezio(lambda x:np.log(x),1.0,1.6,4)
print(f"defTraz: {defTraz}")

# erro
f= lambda x:np.log(x)
i = si.quad(f, 1.0, 1.6)
print (i)

erro = np.abs(I - i[0])
erro
```

```python
# Resoluçao
# Define function to integrate
def f(x):
    return (x + 3) / ((x ** 2) - 1)

# Implementing Simpson's 1/3 
def simpson(x0,xn,n):
    # calculating step size
    h = (xn - x0) / n
    
    # Finding sum 
    integration = f(x0) + f(xn)
    
    for i in range(1,n):
        k = x0 + i*h
        
        if i%2 == 0:
            integration = integration + 2 * f(k)
        else:
            integration = integration + 4 * f(k)
    
    # Finding final integration value
    integration = integration * h/3
    
    return integration

result = simpson(2, 10, 4)
print("Integracao resultado por Simpson é: %0.6f" % (result) )
```
